```python
# @title
import pandas as pd

# Import file Excel untuk dijadikan dataframe
df = pd.read_excel('/content/Real estate valuation data set.xlsx')
df.head()
```

```python
# Merubah nama pada kolom data untuk menjadi lebih singkat
df.rename(columns={"X1 transaction date":"transaction date",
                   "X2 house age":"house age",
                   "X3 distance to the nearest MRT station":"distance to MRT",
                   "X4 number of convenience stores":"convenience store",
                   "X5 latitude":"latitude",
                   "X6 longitude":"longitude",
                   "Y house price of unit area":"price"}, inplace=True)
df
```

Penjelasan:
1. X1 = transaction date = waktu terjadinya pembayaran pembelian rumah
2. X2 = house age = umur dari rumah tersebut sejak dibangun (tahun)
3. X3 = distance to MRT = menyatakan jarak transportasi umum MRT dari lokasi rumah tersebut (meter)
4. X4 = convenience store = jumlah pasar swalayan yang berada di sekitar rumah tersebut
5. X5 = latitude = koordinat rumah tesebut dalam sumbu utara-selatan bumi
6. X6 = longitude = koordinat rumah tesebut dalam sumbu timur-barat bumi
7. Y = price = harga rumah (NT$)

```python
# Melakukan Multiple Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

x = df.drop(columns=['No','price'])
y = df['price']

# Split data untuk training dan testing dengan porsi 90:10
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.1, random_state=4)

# Cek shape dari data training dan testing
print(x_train.shape)
print(y_train.shape)
print(x_test.shape)
print(x_test.shape)
```

```python
# Membuat object linear regression
l_r = LinearRegression()

# Training model menggunakan training data yang sudah di-split
l_r.fit(x_train, y_train)

print(l_r.coef_)
print(l_r.intercept_)

# Visualisasikan nilai koefisien variabel X dan intercept ke dalam dataframe
coef_dict ={
    'features': x.columns,
    'coef_value': l_r.coef_
}
coef = pd.DataFrame(coef_dict, columns=['features','coef_value'])
coef
```

Dari analisis di atas, didapatkan hasil persamaan regresi dari dataset tersebut menjadi Y = -9657.582 + 2.971X1 - 0.276X2 - 0.004X3 + 0.980X4 + 251.790X5 - 21.118X6

```python
# Cek akurasi model yang sudah dibuat
y_pred = l_r.predict(x_test)
l_r.score(x_test, y_test)
```

Model ini mendapatkan nilai akurasi sebesar 0.597 atau dapat dibulatkan menjadi 60%
